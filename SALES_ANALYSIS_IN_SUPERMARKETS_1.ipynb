```python
import pandas as pd

df=pd.read_csv('https://raw.githubusercontent.com/Muskan12m/cutm/master/new_data.csv')
df.head()
#CLEANING OF DATA
nan_df=df[df.isna().any(axis=1)]
nan_df.head()
df = df.dropna(how="any")
df.dropna()


```

```python
df['Quantity Ordered']=pd.to_numeric(df['Quantity Ordered'])
df['Price Each']=pd.to_numeric(df['Price Each'])
df
```

```python
df['Date']=pd.to_datetime(df['Date'])
df
```

```python
#ADDING COLUMN MONTH
df['Month'] = df['Date'].dt.month
df['Month'] = df['Month'].astype('int64')
df.tail()
```

```python
# adding sales column
df['Sales']=df['Quantity Ordered']*df['Price Each']
df.head()
```

```python
#what was the best month for sales? And how much was earned that month?
results=df.groupby('Month').sum()
results
```

```python
# which city has highest no of sales
def get_city(address):
    return address.split(',')[1]
def get_state(address):
    return address.split(',')[2].split(' ')[1]
df["City"]=df['Purchase Address'].apply(lambda x: get_city(x)+' ('+get_state(x)+')')
df.head()
```

```python
results=df.groupby('City').sum()
results
```

```python
import matplotlib.pyplot as plt
import numpy as np
cities= [city for city,df in df.groupby('City')]
plt.scatter(cities,results['Sales'],color="purple",s=100)
plt.xticks(cities,rotation='vertical',size=10)
plt.xlabel('CITY NAME')
plt.ylabel("SALES IN USD($)")
plt.show()
```

```python
#what time should we display advertisements to maximise likelihoodof customer'sbuying product
df['Date']=pd.to_datetime(df['Date'])
df["Hour"]=df["Date"].dt.hour
df["Minute"]=df["Date"].dt.minute
df["Count"]=1
df.head()
```

```python
city= [city for city,df in df.groupby('City')]
print(city)
plt.plot(city,df.groupby(["City"]).count(),color="red")
plt.xticks(city,rotation='vertical')
plt.xlabel("City")
plt.ylabel("NO OF ORDERS")
plt.grid()
plt.title('CITY BASED ANALYSIS OF THE NO. OF ORDERS PLACED')
plt.show()
```

```python

df.head()
branch_group = df.groupby('Branch')
quantity_ordered = branch_group.sum()['Quantity Ordered']
branch = [branch for branch, df in branch_group]
plt.bar(branch, quantity_ordered,color='orange')
plt.xticks(branch, rotation='horizontal', size=10)
plt.xlabel("BRANCH")
plt.ylabel("NO. OF ORDER")
plt.legend(quantity_ordered)
plt.title('NO. OF ORDERS PLACED BRANCHWISE')

plt.show()
```

```python
product_group = df.groupby('Product')
products = [product for product, df in product_group]
quantity_ordered = product_group.sum()['Quantity Ordered']
prices=df.groupby("Product").mean()["Price Each"]
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color='g',)
ax2.plot(products, prices, color='black')
ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered',color='g')
ax2.set_ylabel('Price ($)', color='g')
ax1.set_xticklabels(products, rotation='vertical')
plt.title('PRODUCT SALE ON THE BASIS OF ITS PRICE AND NO. OF ORDERS')
plt.legend(prices,bbox_to_anchor =(1, 1))


fig.show()
```
